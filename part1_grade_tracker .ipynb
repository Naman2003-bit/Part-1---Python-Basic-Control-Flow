# ***Task 1 — Data Parsing & Profile Cleaning***

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]

# To store the cleaned students
students = []
for s in raw_students:
    # To clean the names
  clean_name = s["name"].strip().title()
    ## strip()- Removes the extra spaces
    ## title()- Convert first letter capital

    # Convert the roll strings into Integer
  clean_roll = int(s["roll"])

    # convert the marks_str into list
  clean_marks = [int(x) for x in s["marks_str"].split(", ")]

    # Name validation
  words = clean_name.split()
  is_valid = True
  for word in words:
    if not word.isalpha():
      is_valid = False
  if is_valid:
        status = "Valid name"
  else:
        status ="Invalid name"

    # Now creating clean dictionary
  cleaned = {
      "name": clean_name,
      "roll": clean_roll,
      "marks": clean_marks,
    }
  students.append(cleaned)
  print("=============================")
  print(f"Student : {clean_name}  {status}")
  print(f"Roll : {clean_roll}")
  print(f"Marks : {clean_marks}")
  print("=============================")


#Printing the name of Roll 103 in All Caps and Lowercase
for s in students:
  if s["roll"] == 103:
    print(f"All CAPS: {s['name'].upper()}")
    print(f"Lowercase: {s['name'].lower()}")
```

# ***Task 3 — Class Performance Summary***

```python
from os import name
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

# Things we need to check for each student
pass_count = 0
fail_count = 0
topper_name = ""
topper_avg = 0
all_averages = []

#Print the Header of table
print(f"{'Name':<18} | {'Average': <8} | Status")
print("-" * 40)

for student in class_data:
  #Element of tuple
    name = student[0]
    marks = student[1]

#Average Calculation
    average = round(sum(marks)/ len(marks),2)

#Check whether it's pass or fail
    if average >= 60:
     status = "Pass"
     pass_count += 1
    else:
      status = "Fail"
      fail_count += 1

#Tracker Topper
    if average > topper_avg:
       topper_avg = average
       topper_name = name

# Need to stroe all the averages for class average
    all_averages.append(average)

#Printing Row (<18 mean 18 character's width left aligned)
    print(f"{name:<18} | {average:<8} | {status}")

#Summary
print("-" * 40)
print(f"Pass : {pass_count} Students")
print(f"Fail : {fail_count} Students")
print(f"Topper : {topper_name} with {topper_avg}")


#Class Average
class_avg = round(sum(all_averages) / len(all_averages),2)
print(f"Class Average : {class_avg}")
```

# ***Task 4 — String Manipulation Utility***

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# Striping Remove the extra spaces
clean_essay = essay.strip()
print("Step 1 - Stripped: ")
print(clean_essay)
print()

#Conversion
print("Step 2 - Title Case: ")
print(clean_essay.title())
print()

#Count the number of "python" Word
print("Step 3 - Python Word Count: ")
print(clean_essay.count("python"))
print()

#Convert "python" to "Python 🐍"
print("Step 4 - Python to Python 🐍: ")
print(clean_essay.replace("python", "Python 🐍"))
print()

#Spliting into sentences into ("- ")
sentences = clean_essay.split(". ")
print("Step 5 - Splitting into Sentences: ")
print(sentences)
print()

# Print each sentence numbered, ensuring it ends with a period
print("Step 6 -  Numbered Sentences: ")
for i, sentences in enumerate(sentences, 1):
  if not sentences.endswith("."):
    sentences += "."
  print(f"{i}. {sentences}")
print()
```
